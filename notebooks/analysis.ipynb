```python
import geopandas as gpd
import matplotlib.pyplot as plt
import fiona
from shapely.geometry import Polygon
# Create a sample GeoDataFrame with polygon geometries
gdb_path = '/Volumes/HOLMES/NITENI/github/nb_001_batam_atlas/data/boundaries/KOTA BATAM/RBI50K_KOTABATAM_KUGI50_20221231.gdb'
fiona.listlayers(gdb_path)  # List layers in the shapefile
```

```python
batas_kecamatan = gpd.read_file(
    gdb_path, layer='ADMINISTRASI_AR_KECAMATAN'
)
batas_kecamatan.columns
batas_kecamatan.plot()
plt.show()

batas_kecamatan.head()
len(batas_kecamatan)
batas_kecamatan.shape

batas_kecamatan.head(27)
```

```python
batas_kecamatan = gpd.read_file(
    gdb_path, layer='ADMINISTRASI_AR_KECAMATAN'
)
bound_kecamatan = batas_kecamatan[
    batas_kecamatan['WADMKK'] == "Kota Batam"
].copy()

print(bound_kecamatan.shape)
print(bound_kecamatan.total_bounds)
print(bound_kecamatan.crs)

bound_kecamatan.to_crs("EPSG:4326", inplace=True)
bound_kecamatan.to_file(
    "../data/boundaries/KOTA BATAM/batas_kecamatan.geojson", 
    driver="GeoJSON"
)
```
